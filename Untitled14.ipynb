```python
def outlier_count(col, data=df):
    print(15*'-' + col + 15*'-')
    q75, q25 = np.percentile(data[col], [75, 25])
    iqr = q75 - q25
    min_val = q25 - (iqr*1.5)
    max_val = q75 + (iqr*1.5)
    outlier_count = len(np.where((data[col] > max_val) | (data[col] < min_val))[0])
    outlier_percent = round(outlier_count/len(data[col])*100, 2)
    print('Number of outliers: {}'.format(outlier_count))
    print('Percent of data that is outlier: {}%'.format(outlier_percent))
for col in columns_dict:
    outlier_count(col)
```

```python
mean_developing = developing['winsorized_Life_Expectancy'].mean()
mean_developed = developed['winsorized_Life_Expectancy'].mean()

print(f"The mean life expectancy of developing countries: {mean_developing}")
print(f"The mean life expectancy of developed countries: {mean_developed}")

s1 = developing['winsorized_Life_Expectancy'].std()
s2 = developed['winsorized_Life_Expectancy'].std()


n1 = developing['winsorized_Life_Expectancy'].count()
n2 = developed['winsorized_Life_Expectancy'].count()
print(f"S1 is: {s1} while s2 is: {s2}")
print("..............")
import math


sp = math.sqrt(((n1-1)*s1**2+(n2-1)*s2**2)/(n1+n2-2)) 

print(f"The pooled estimate of the commom stdev is: {round(sp,3)}, which is a value between {round(s1,3)} and {round(s2,3)}")
```

```python
z = (mean_developing-mean_developed)/(sp*math.sqrt((1/n1)+(1/n2)))
z
```

```python
from scipy import stats
from statsmodels.stats import weightstats as stests

# computing the z_test and p-value
ztest ,pval1 = stests.ztest(developing['winsorized_Life_Expectancy'], developed['winsorized_Life_Expectancy'], value=0,alternative='two-sided')
print(ztest, pval1)
if pval1 < 0.05:
    print("reject null hypothesis and accept the alternate hypothesis")
else:
    print("accept null hypothesis")
```
